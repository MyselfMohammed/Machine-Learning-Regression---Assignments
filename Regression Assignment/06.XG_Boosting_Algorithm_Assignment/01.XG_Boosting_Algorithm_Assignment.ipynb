```python
import pandas as pd
```

```python
dataset = pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset = pd.get_dummies(dataset,dtype=int, drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
independent
```

```python
dependent = dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
! pip install xgboost
```

```python
#Model creation using XG Boost Algorithm
#XG Boosting Works out in Deep Level based on any one Threshold Value Like Age = 50 and comaparing with Output
import xgboost as xgb
regressor = xgb.XGBRegressor()

regressor=regressor.fit(X_train,Y_train)
```

```python
#prediction
#Keyword - predict is function of Application 
Y_pred=regressor.predict(X_test)
```

```python
# Evaluation Metrics - Calculate R-Square Value
from sklearn.metrics import r2_score
r_score = r2_score(Y_test,Y_pred)
r_score
```

```python
# Naming the Model with its extension .sav
import pickle
filename = "finalized_model_XG_Boosting_Algorithm_Assignment.sav"

#Model will be saved in respective path, once after the below Code Execution
# Keyword - wb refers to "write binary"
pickle.dump(regressor,open(filename,'wb'))
```

```python
# Load the Saved Model and check the same to send it for Phase-02 (Deployment)
# # Keyword - rb refers to "read binary"

loaded_model=pickle.load(open("finalized_model_XG_Boosting_Algorithm_Assignment.sav",'rb'))
result = loaded_model.predict([[10000,20000,3000,1,0]]) 
result
```
