```python
import pandas as pd
```

```python
dataset = pd.read_csv("insurance_pre.csv")
dataset
```

```python
#Observe the Categorical Column and finalize whether it is Nominal or Ordinal Value
#Covert the Categorical Column (Here Nominal) into Number Values (One Hot Encoding / Column Expansion)

dataset = pd.get_dummies(dataset, dtype=int, drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
independent
```

```python
dependent = dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
#Execute the First Two Procedures which only differs Model to Model
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor (n_estimators = 50, random_state = 0, criterion = 'absolute_error', max_features = 'log2')

regressor = regressor.fit(X_train,Y_train)
#Once the above steps Executed a New Model would be created which is Nothing but an AI Brain
```

```python
#prediction
#Keyword - predict is function of Application 
Y_pred=regressor.predict(X_test)
```

```python
# Evaluation Metrics - Calculate R-Square Value
from sklearn.metrics import r2_score
r_score = r2_score(Y_test,Y_pred)
r_score
```

```python
# Naming the Model with its extension .sav
import pickle
filename = "finalized_model_Random_Forest_Overall_Assignment.sav"

#Model will be saved in respective path, once after the below Code Execution
# Keyword - wb refers to "write binary"
pickle.dump(regressor,open(filename,'wb'))
```

```python
# Load the Saved Model and check the same to send it for Phase-02 (Deployment)
# # Keyword - rb refers to "read binary"

loaded_model=pickle.load(open("finalized_model_Random_Forest_Overall_Assignment.sav",'rb'))
result = loaded_model.predict([[20,30,1,1,0]]) 
result
```
