```python
# pandas is a Library and it contains the functions related to database
# 1. Create a New File with Name - 01.Simple_Linear_Regression.py
# 2.Import an Existing Library - pandas
# (The Library - pandas, has all the inbuilt functions those handles Tables and Databases)
# 3. Assign the Imported Library - pandas (Using a Keyword - as) to any Short Variable (Here, We assigned as "pd") for ease of access

import pandas as pd
```

```python
# dataset is called by read_csv function under pandas and assigned to the Variable - dataset
# Method : dataset = className.functionName("Enter the Parameter")
# Method Used Here : pd.read_csv("Salary_Data")
# NOTE : Dataset (File) should be available in the same location of Python Program which is to be Executed

dataset = pd.read_csv ("Salary_Data.csv")
dataset
```

```python
# Split the Dependent and Independent Variables
# Input
# Method : variableName = alreadyAssignedVariableName ("columnName")
# Method Used Here : independent=dataset[["Years Of Experience"]]

independent=dataset[["Years_Of_Experience"]]
independent
```

```python
# Split the Dependent and Independent Variables
# Output
# Method : variableName = alreadyAssignedVariableName ("columnName")
# Method Used Here : dependent=dataset[["Salary"]]

dependent=dataset[["Salary"]]
dependent
```

```python
# 1. Splitting Test Set
# 1.Import a Particular Function from an Existing Library - sklearn.model_selection
# (Note : All the Algorithms of Machine Learning is provided by sklearn) 
# Method Used Here : from sklearn.model_selection import train_test_split
# 2.train_test_split should be assigned with 4 variables as X_train, X_test, Y_train and Y_test, since it gives 4 Results

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
# Model Creation/Learning - Phase
# --> This is the actual Learning Phase
# 1.Import a Particular Function from an Existing Library - sklearn.linear_model
# Method Used Here : from sklearn.linear_model import LinearRegression
# 2.LinearRegression (It is a Class) should be assigned to a New Variable - regressor
# 3. Call a Particular Function from an Existing Class with Parameters
# Method Used Here : from sklearn.linear_model import LinearRegression
# regressor = LinearRegression()
# regressor.fit(X_train,Y_train)

from sklearn.linear_model import LinearRegression 
# --> Procedure-01 To Calculate Both Weight and Bais

regressor = LinearRegression()
# --> Procedure-02 To Calculate Both Weight and Bais

regressor.fit(X_train,Y_train)

# NOTE : Model is ready where Value of Both Weight and Bais is Calculated, if the above set of Procedure/Block is Executed
```

```python
#To Calculated the Value of Weight
weight=regressor.coef_
weight
```

```python
#To Calculated the Value of Bais
bais = regressor.intercept_
bais
```

```python
#prediction
#Keyword - predict is function of Application 
Y_pred=regressor.predict(X_test)
```

```python
# Evaluation Metrics - Calculate R-Square Value
from sklearn.metrics import r2_score
r_score = r2_score(Y_test,Y_pred)
r_score
```

```python
# Naming the Model with its extension .sav
import pickle
filename = "finalized_model_linear.sav"
```

```python
#Model will be saved in respective path, once after the below Code Execution
# Keyword - wb refers to "write binary"
pickle.dump(regressor,open(filename,'wb'))
```

```python
# Load the Saved Model and check the same to send it for Phase-02 (Deployment)
# # Keyword - rb refers to "read binary"

loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
result = loaded_model.predict([[11]]) 
result
```
